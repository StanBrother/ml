```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import seaborn as sns
```

```python
df = pd.read_csv("./data.csv")
df
```

```python
sns.regplot(df, x="area", y="price")
```

```python
reg = LinearRegression()
reg.fit(df[["area"]], y=df["price"])
```

```python
reg.predict([[3600]])
```
